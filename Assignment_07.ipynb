## Hierarchy Clustering

```python
n = MinMaxScaler()
data1= n.fit_transform(crime.iloc[:,1:].to_numpy())
crimes = pd.DataFrame(data1, columns = crime.columns[1:])
crimes.head(4)
```

```python
# Normalization function 
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)

# Normalized data frame (considering the numerical part of data)
crime_norm = norm_func(crime.iloc[:,1:])

crime_norm
```

### Clustering 

```python
# create dendrogram
fig = plt.figure(figsize=(15,8))
dendrogram = sch.dendrogram(sch.linkage(crimes, method='average'))
```

```python
# create dendrogram
fig = plt.figure(figsize=(15,8))
dendrogram = sch.dendrogram(sch.linkage(crimes, method='ward'))
```

```python
# create clusters
hc1 = AgglomerativeClustering(n_clusters=4, affinity = 'euclidean', linkage = 'average')
```

```python
# save clusters for chart
y2 = hc1.fit_predict(crimes)
cc = pd.DataFrame(y2,columns=['Clusters'])
cc.head(6)
```

```python
crimes2 = pd.concat([crime,cc],axis=1)
crimes2.head(7)
```

```python
crimes2.sort_values("Clusters").reset_index()
```

```python
crimes2['Clusters'].value_counts()
```

## K-Means

```python
kmeans = KMeans(n_clusters=4,random_state=0)
kmeans.fit(crimes)
```

```python
kmeans.inertia_
```

```python
wcss = []
for i in range(1, 8):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(crimes)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1, 8), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```
